Bài 4.3 

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

# Tạo dataset
X, y = make_classification(
    n_samples=200,
    n_features=10,
    random_state=42
)

# Chia train/test
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42
)

print("=== Quan sát Overfitting ===\n")

# Thử nhiều độ sâu khác nhau
for depth in [1, 2, 3, 5, 10, None]:

    # Tạo model
    model = DecisionTreeClassifier(
        max_depth=depth,
        random_state=42
    )

    # Train model
    model.fit(X_train, y_train)

    # Accuracy
    train_acc = model.score(X_train, y_train)
    test_acc = model.score(X_test, y_test)

    # In kết quả
    print(f"depth={depth}")
    print(f"Train Accuracy = {train_acc:.3f}")
    print(f"Test Accuracy  = {test_acc:.3f}")
    print("-" * 35)
```

Fill: Quan sát: Khi depth tăng, train accuracy và test accuracy thay đổi như thế nào?
Khi depth tăng, train accuracy thường tăng liên tục và có thể đạt gần 100%. Test accuracy lúc đầu tăng nhưng sau đó giảm hoặc không tăng nữa vì model bắt đầu học quá chi tiết dữ liệu train.

Fill: Depth nào gây overfitting rõ nhất? Bạn nhận ra bằng dấu hiệu gì?
Depth = None hoặc depth rất lớn như 10 thường gây overfitting rõ nhất. Dấu hiệu là train accuracy rất cao (gần 100%) nhưng test accuracy thấp hơn nhiều, chứng tỏ model nhớ dữ liệu train nhưng tổng quát kém.

Fill: Bạn sẽ chọn depth nào để deploy? Tại sao?
Tôi sẽ chọn depth khoảng 3 hoặc 5 vì train accuracy và test accuracy cân bằng hơn, chênh lệch nhỏ và model có khả năng tổng quát tốt trên dữ liệu mới.
